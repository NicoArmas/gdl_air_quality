```python
#problemi:
# batch size diverse (quella la scartiamo)
# time window non sufficiente (forse non succede)


window = 12
batch_size = 32
n = len(aq.dataframe())
n_batches = n // (window * batch_size)
tens = None
i = 0
rows = 0
count_rows = 0
batches = []


#for index, row in aq.dataframe().iterrows():
while True:
    batch = None
    #i = rows + 1
    #count_rows = 1
    while batch is None or batch.shape[-1] < batch_size:
        tens = torch.FloatTensor(list(aq.dataframe().iloc[rows]))
        tens = tens.view(-1, len(tens))
        i = rows + 1
        count_rows = 1
        while count_rows <= window:
            tmp = aq.dataframe().iloc[i]
            tmp = torch.FloatTensor(list(tmp.copy()))
            tmp = tmp.view(-1, len(tmp))
            tens = torch.cat([tens, tmp], 0)
            i += 1
            count_rows += 1
            
        tens = tens.view(tens.shape[0], tens.shape[1], -1)

        if batch is None:
            batch = torch.clone(tens)
        else:
            batch = torch.cat([batch, tens], 2)
        rows += 1

    #rows += 1
    batches.append(torch.clone(batch))
    if len(batches) == n_batches:
        break

file_to_store = open("LSTM_batches.pickle", "wb")
pickle.dump(batches, file_to_store)
file_to_store.close()
```

```python
file_to_read = open("LSTM_batches.pickle", "rb")
loaded_object = pickle.load(file_to_read)
file_to_read.close()
```
